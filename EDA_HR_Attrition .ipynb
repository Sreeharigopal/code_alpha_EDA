# HR Attrition Dataset
The HR Attrition Dataset is a commonly used dataset in Human Resources analytics to understand the factors that contribute to employee attrition (i.e., employees leaving a company).

### Key Features of the Dataset

Each row in the dataset represents an individual employee, and the columns contain various infomation stated below


| Feature Name      | Description                                                                 |
| ------------------| --------------------------------------------------------------------------- |
| Attrition         | Whether the employee left the company (`Yes` or `No`) – target variable     |
| Age               | Age of the employee                                                         |
| BusinessTravel    | Frequency of business travel (e.g., Rarely, Frequently)                     |
| Department        | Employee’s department (e.g., Sales, R\&D, HR)                               |
| DistanceFromHome  | Distance in kilometers from home to office                                  |
| Education         | Education level (1–5 scale)                                                 |
| Gender            | Male/Female                                                                 |
| JobRole           | Role/position of the employee (e.g., Sales Executive, Research Scientist)   |
| JobSatisfaction   | Satisfaction level (1–4 scale)                                              |
| MonthlyIncome     | Monthly salary of the employee                                              |
| OverTime          | Whether the employee works overtime (`Yes` or `No`)                         |
| TotalWorkingYears | Total years of professional experience                                      |
| YearsAtCompany    | Years spent at the current company                                          |
| WorkLifeBalance   | Rating of work-life balance (1–4 scale)                                     |

### Identify Trends, Patterns, and Anomalies

### a. Attrition Distribution

```python
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title('Attrition Count')
plt.show()
```

### b. Attrition by Department

```python
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('Attrition by Department')
plt.xticks(rotation=45)
plt.show()
```

### c. Age Distribution by Attrition

```python
sns.boxplot(data=df, x='Attrition', y='Age')
plt.title('Age vs Attrition')
plt.show()
```

### d. Monthly Income vs Attrition

```python
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income vs Attrition')
plt.show()
```

### e. Overtime vs Attrition

```python
sns.countplot(data=df, x='OverTime', hue='Attrition')
plt.title('OverTime vs Attrition')
plt.show()
```

# Test Hypotheses / Validate Assumptions

### (a). Is there a significant difference in average income between attrition groups?

```python
from scipy.stats import ttest_ind

group_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome']
group_no = df[df['Attrition'] == 'No']['MonthlyIncome']

t_stat, p_val = ttest_ind(group_yes, group_no)
print(f"T-test results: t-statistic = {t_stat:.2f}, p-value = {p_val:.4f}")
```

### (b). Correlation heatmap (for numeric features)

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
